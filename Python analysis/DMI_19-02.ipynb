```python
import pandas as pd
```

```python
df = pd.read_csv('dk_daily_27080_5165_601.csv', delimiter = ';', na_values='null')
df
```

```python
#välja ut maxima för varje år

df['Date'] = pd.to_datetime(df['year'])
df['elem_val'] = df['elem_val'].str.replace(',','.')
df['elem_val'] = pd.to_numeric(df['elem_val'])


max_per_year = df.loc[df.groupby(df['year'])['elem_val'].idxmax()]
max_per_year =  max_per_year.drop(columns=['month', 'day', 'Date', 'hour'])
max_per_year
```

```python
max_per_year.to_csv('DMIRep19-02/maximi/TranebjergOst_5165_1874_2019_maxi.csv', index = False)
```

```python
max_per_year.plot(x = 'year', y = 'elem_val')
```
