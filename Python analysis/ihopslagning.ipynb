```python
import pandas as pd
```

```python
df1 = pd.read_csv('komplettering_tr12-06/NörreVorupör_2012_2025_max.csv', index_col = 'observed')
df2 = pd.read_csv('tr12-06_1960-2011/NorreVurpör.csv', names = ['year', 'val'], index_col = 'year')

df2['val'] = df2['val'].str.replace(',','.')
df2['val'] = pd.to_numeric(df2['val'])
df1 = df1.rename(columns = {'precip_past24h' : 'val'})
df2 = df2.rename(columns = {'observed' : 'year', 'precip_past24h' : 'precipitation'})

df3 = pd.concat([df2, df1])
df3
```

```python
df3.to_csv('DMIRep19-02/max_hela _tidsspannet/NörreVorupör_max.csv')
```
