```r
qq_plot <- function(mdl) {


    par <- mdl$results$par
    
    location_trend <- mdl$par.models$location[[2]]
    scale_trend <- mdl$par.models$scale[[2]]

    data <- mdl$cov.data$rain
    cov <- mdl$cov.data$GMST

    
  # Hämta formparameter (alltid konstant)
    xi_hat <- par["shape"]

    if (location_trend == "COV") {
        mu0  <- par["mu0"]
        mu1  <- par["mu1"]
        mu_hat    <- mu0 + mu1 * cov
    }
    
    else {
            mu0  <- par["mu0"]
            mu_hat    <- rep(mu0, length(data))
        }

    if (scale_trend == "COV") {
        phi0 <- par["phi0"]
        phi1 <- par["phi1"]
        sigma_hat <- exp(phi0 + phi1 * cov)
        }

    else {
        phi0 <- par["log.scale"]
        sigma_hat <- rep(exp(phi0), length(data))
        }
    
  # Sortera data och motsvarande parametrar efter datastorlek
    ord       <- order(data)
    x_sorted  <- data[ord]
    mu_sorted <- mu_hat[ord]
    si_sorted <- sigma_hat[ord]
    m         <- length(x_sorted)
  
  # Beräkna residualer
  x_hat <- (1/xi_hat) * log(1 + xi_hat * (x_sorted - mu_sorted) / si_sorted)
  
  # Beräkna teoretiska Gumbel-kvantiler
  i <- 1:m
  Q <- -log(-log(i / (m + 1)))

    # Plotta
    plot(Q, x_hat,
         xlab = "Teoretiska kvantiler",
         ylab = "Empiriska residualer",
         main = "Gumbel QQ-plot")
    abline(0, 1, col = "red")
}
```
