```r
qq_plot <- function(mdl) {

    par <- mdl$results$par
    data <- mdl$cov.data$rain
    cov <- mdl$cov.data$GMST

    
  # Hämta formparameter (alltid konstant)
    xi_hat <- par["shape"]
  
  # Beräkna tidsberoende mu och sigma beroende på modell
    location_trend <- mdl$par.models$location[[2]]
    scale_trend <- mdl$par.models$scale[[2]]
    
    if (location_trend == "GMST" & scale_trend == "GMST") {
        title = "Trend i mu & sigma"
        
        # Trend i både mu och sigma
        mu0  <- par["mu0"]
        mu1  <- par["mu1"]
        phi0 <- par["phi0"]
        phi1 <- par["phi1"]
    
        mu_hat    <- mu0 + mu1 * cov
        sigma_hat <- exp(phi0 + phi1 * cov)
        
  } else if (scale_trend == "GMST") {
        # Endast trend i sigma
        title = "Trend i sigma"
        
        mu0  <- par["location"]
        phi0 <- par["phi0"]
        phi1 <- par["phi1"]
        
        mu_hat    <- rep(mu0, length(data))
        sigma_hat <- exp(phi0 + phi1 * cov)
    
  } else if (location_trend == "GMST") {
        # Endast trend i mu

        title = "Trend i mu"
        mu0  <- par["mu0"]
        mu1  <- par["mu1"]
        phi0 <- par["log.scale"]
        
        mu_hat    <- mu0 + mu1 * cov
        sigma_hat <- rep(exp(phi0), length(data))
    
  } else {
        # Stationär modell

        title = "Stationär"
        mu0  <- par["location"]
        phi0 <- par["log.scale"]
        
        mu_hat    <- rep(mu0, length(data))
        sigma_hat <- rep(exp(phi0), length(data))
  }
  
    # Beräkna residualer enligt ekvation (9)
    x_hat_unsorted <- (1/xi_hat) * log(1 + xi_hat * (data - mu_hat) / sigma_hat)
    x_hat <- sort(x_hat_unsorted)
  
    # Beräkna teoretiska Gumbel-kvantiler enligt ekvation (10)
    m <- length(data)
    i <- 1:m
    
    Q <- -log(-log(i / (m + 1)))


      # Plotta
    plot(Q, x_hat,
         xlab = "Teoretiska kvantiler",
         ylab = "Empiriska residualer",
         main = "QQ-plot")
    abline(0, 1, col = "red")

    mtext(title, 
      side = 3,      # 3 = ovanför plotten
      line = 0,      # 0 = precis under huvudtiteln
      cex = 1)     # Lite mindre text

    #return (list(diff(range(mu_hat)), diff(range(sigma_hat))))
}
```

```r
options(repr.plot.width = 9, repr.plot.height = 9)
par(oma = c(0, 0, 2, 0))  # Yttre marginal ovanför
par(mfrow = c(2, 2))

qq_plot(mdl_Nordby2_stat)
qq_plot(mdl_Nordby2_1)
qq_plot(mdl_Nordby2_2)
qq_plot(mdl_Nordby2_3)

mtext("Datamängd Nordby", outer = TRUE, cex = 1.2, font = 2)
```
